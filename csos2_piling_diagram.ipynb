```python
%pylab inline

import matplotlib.gridspec as gridspec
import numpy as np
import pandas as pd
import pylab
import seaborn
```

```python
# Read data from CSV. Thank you Pandas!
df_130 = pd.read_csv("data/Csos2B_peptide_counts.csv")
df_70 = pd.read_csv("data/Csos2A_peptide_counts.csv")
```

```python
csos2B_counts = np.zeros(1000)
csos2A_counts = np.zeros(1000)

# For each AA in a peptide, add 1 to its count to make piles.
for start, stop in zip(df_130.Start, df_130.Stop):
    if not np.isfinite(start) or not np.isfinite(stop):
        continue
    csos2B_counts[int(start):int(stop+1)] += 1
# Now for S2A.
for start, stop in zip(df_70.Start, df_70.Stop):
    if not np.isfinite(start) or not np.isfinite(stop):
        continue
    csos2A_counts[int(start):int(stop+1)] += 1
    
# Normalize to max count for both forms so that 
# plots are to the same scale (0.0 to 1.0). 
csos2B_counts_norm = csos2B_counts / csos2B_counts.max()
csos2A_counts_norm = csos2A_counts / csos2A_counts.max()
```

```python
gs = gridspec.GridSpec(2, 1, height_ratios=[1, 1])

# End of S2A plus 1 for graphical purposes.
fs_aa = 568

# Colors for plotting. 
S2B_color = seaborn.xkcd_rgb['teal']
S2A_color = seaborn.xkcd_rgb['mustard']
dark_grey = seaborn.xkcd_rgb['dark grey']

# Figure.
pylab.figure(figsize=(10,10))
ax0 = pylab.subplot(gs[0])

# Plot piling diagram for S2A.
ax0.plot(np.arange(1, 1001), csos2A_counts_norm, color=S2A_color, lw=1, label='Csos2A')
ax0.fill_between(np.arange(1, 1001), csos2A_counts_norm, color=S2A_color, alpha=0.8)

# Line & text marking end of S2A.
ax0.axvline(fs_aa, color=dark_grey, ls='--')
ax0.text(fs_aa + 1, 0.9, "AA %d" % (fs_aa - 1), rotation=270, fontsize=14, color=dark_grey, fontweight="bold")

# Axis limits and ticks. 
ax0.set_ylim(0, 1.05)
ax0.set_xlim(1, 880)
ax0.xaxis.set_ticklabels([])

# Plot piling diagram for S2B.
ax1 = pylab.subplot(gs[1])
ax1.plot(np.arange(1, 1001), csos2B_counts_norm, color=S2B_color, lw=1, label='Csos2B')
ax1.fill_between(np.arange(1, 1001), csos2B_counts_norm, color=S2B_color, alpha=0.8)

# Line marking end of S2A.
ax1.axvline(fs_aa, color=dark_grey, ls='--')

# Axis limits same as above.
ax1.set_ylim(0, 1.05)
ax1.set_xlim(1, 880)

# Axis labels.
ax1.set_xlabel('Amino Acid Position', fontsize=16)
ax1.set_ylabel('Fraction of Peptide Counts', fontsize=16)

# Ticks only on bottom plot. 1-indexed for protein people.
xticks = np.arange(0, 900, 100)
xticks[0] = 1
ax1.set_xticks(xticks)

# Make ticks larger for both axes.
ax0.tick_params(axis='both', which='major', labelsize=12)
ax1.tick_params(axis='both', which='major', labelsize=12)

# Tighten up spacing between subplots.
pylab.subplots_adjust(hspace=0.05)

pylab.savefig('figs/csos2_piling.svg')
pylab.savefig('figs/csos2_piling.png')
pylab.show()
```
